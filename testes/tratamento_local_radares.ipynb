# Verificação de muitos acidentes e poucos radares

```python
%pip install geopandas
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from sklearn.neighbors import NearestNeighbors

def preparar_radares(df1, df2):
    # Seleciona latitude e longitude apenas
    df1 = df1[['latitude', 'longitude']].dropna()
    df2 = df2[['latitude', 'longitude']].dropna()
    return pd.concat([df1, df2], ignore_index=True)

# Função principal para análise de proximidade
def analisar_proximidade(df_acidentes, df_radares, raio_metros=1000):
    # GeoDataFrames
    gdf_acidentes = gpd.GeoDataFrame(df_acidentes, 
                                     geometry=gpd.points_from_xy(df_acidentes['longitude'], df_acidentes['latitude']),
                                     crs="EPSG:4326")
    gdf_radares = gpd.GeoDataFrame(df_radares, 
                                   geometry=gpd.points_from_xy(df_radares['longitude'], df_radares['latitude']),
                                   crs="EPSG:4326")

    # Converte para sistema métrico (SIRGAS 2000 / UTM zona 22S para sul do Brasil)
    gdf_acidentes = gdf_acidentes.to_crs("EPSG:31982")
    gdf_radares = gdf_radares.to_crs("EPSG:31982")

    # Calcula distância até radar mais próximo
    X = gdf_acidentes.geometry.apply(lambda p: (p.x, p.y)).tolist()
    Y = gdf_radares.geometry.apply(lambda p: (p.x, p.y)).tolist()
    nn = NearestNeighbors(n_neighbors=1).fit(Y)
    dist, idx = nn.kneighbors(X)
    gdf_acidentes["dist_radar_m"] = dist

    # Filtra acidentes sem radar num raio específico
    sem_radar = gdf_acidentes[gdf_acidentes["dist_radar_m"] > raio_metros]

    return sem_radar, gdf_acidentes
```

```python
import folium
from folium.plugins import MarkerCluster

# Usar uma amostra para criar o mapa se os dados forem grandes
# Neste exemplo, vamos simular a criação do mapa

def gerar_mapa_interativo(gdf_acidentes, gdf_zonas_criticas):
    # Converte para WGS84 para exibir no mapa (latitude/longitude)
    gdf_acidentes = gdf_acidentes.to_crs("EPSG:4326")
    gdf_zonas_criticas = gdf_zonas_criticas.to_crs("EPSG:4326")

    # Centro do mapa
    lat_centro = gdf_acidentes.geometry.y.mean()
    lon_centro = gdf_acidentes.geometry.x.mean()

    mapa = folium.Map(location=[lat_centro, lon_centro], zoom_start=6)

    # Cluster de todos os acidentes
    cluster_acidentes = MarkerCluster(name="Todos os Acidentes").add_to(mapa)
    for _, row in gdf_acidentes.iterrows():
        folium.CircleMarker(
            location=[row.geometry.y, row.geometry.x],
            radius=3,
            color="gray",
            fill=True,
            fill_opacity=0.5,
            popup=f"Dist. radar: {row['dist_radar_m']:.1f} m"
        ).add_to(cluster_acidentes)

    # Zonas críticas (sem radar por perto)
    for _, row in gdf_zonas_criticas.iterrows():
        folium.CircleMarker(
            location=[row.geometry.y, row.geometry.x],
            radius=6,
            color="red",
            fill=True,
            fill_opacity=0.8,
            popup=f"⚠️ Crítico: {row['dist_radar_m']:.1f} m do radar"
        ).add_to(mapa)

    folium.LayerControl().add_to(mapa)
    return mapa

# Exemplo de chamada (com dados reais):
# mapa = gerar_mapa_interativo(acidentes_com_distancia, zonas_criticas)
# mapa.save("mapa_interativo.html")

```

```python

def gerar_tabela_pontos_criticos(gdf_zonas_criticas):
    # Arredonda localização para agrupar
    gdf_zonas_criticas['lat_round'] = gdf_zonas_criticas.geometry.y.round(3)
    gdf_zonas_criticas['lon_round'] = gdf_zonas_criticas.geometry.x.round(3)

    # Agrupa por ponto aproximado + info da rodovia
    tabela = gdf_zonas_criticas.groupby(
        ['uf', 'br', 'km', 'municipio', 'lat_round', 'lon_round']
    ).agg({
        'id': 'count',
        'mortos': 'sum',
        'feridos': 'sum',
        'dist_radar_m': 'mean'
    }).reset_index()

    # Renomeia colunas
    tabela.rename(columns={
        'id': 'qtd_acidentes',
        'dist_radar_m': 'distancia_media_m'
    }, inplace=True)

    # Ordena por acidentes, mortos e distância
    tabela = tabela.sort_values(by=['qtd_acidentes', 'mortos', 'distancia_media_m'], ascending=[False, False, False])
    return tabela
```
