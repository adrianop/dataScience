```python
import pandas as pd
```

```python
df = pd.read_csv('../../dados/acidentes/por ocorrencia/datatran_agrupado.csv', sep=',')
```

```python
df.head(-1)
```

```python
df.shape
```

```python
ufs = ['RS', 'SC', 'PR']
```

```python
df = df[df['uf'].isin(ufs)]
```

```python
import pandas as pd
from dateutil import parser

# Função para tentar converter qualquer formato
def parse_date(x):
    try:
        return parser.parse(x, dayfirst=True)
    except:
        return pd.NaT

# Aplicar a função
df['data_inversa'] = df['data_inversa'].astype(str).apply(parse_date)
```

```python
df.shape
```

```python
df.head(-1)
```

```python
df['dia'] = df['data_inversa'].dt.day
df['mes'] = df['data_inversa'].dt.month
df['ano'] = df['data_inversa'].dt.year
df.head()
```

```python
# Se houve pelo menos 1 morte, grave = 1, senão grave = 0
df['grave'] = df['mortos'].apply(lambda x: 1 if x > 0 else 0)
df.head(50)
```

```python
# Verificar valores nulos
nulls = df[['dia', 'mes', 'ano', 'dia_semana', 'horario', 'uf', 'br', 'km',
            'fase_dia', 'sentido_via', 'condicao_metereologica', 'tipo_pista', 'tracado_via']].isnull().sum()

print("Valores nulos por coluna:")
print(nulls[nulls > 0])
```

```python
df.describe()
```

```python
df['km'] = pd.to_numeric(df['km'], errors='coerce')
df['km'].value_counts()
```

```python
# Contar total de linhas antes
total_antes = len(df)

# Filtrar: manter apenas km > 0
df = df[df['km'] >= 0]

# Contar total depois
total_depois = len(df)

# Calcular eliminadas
eliminadas = total_antes - total_depois

print(f"Linhas eliminadas: {eliminadas}")
print(f"Total restante: {total_depois}")
```

```python
df['br'] = pd.to_numeric(df['br'], errors='coerce')
df['br'].value_counts()
```

```python
df = df[df['fase_dia'].astype(str).str.lower() != '(null)']
df['fase_dia'].value_counts()
```

```python
df = df[df['condicao_metereologica'].astype(str).str.lower() != 'ignorado']
df = df[df['condicao_metereologica'].astype(str).str.lower() != 'ignorada']
df = df[df['condicao_metereologica'].astype(str).str.lower() != '(null)']
df['condicao_metereologica'].value_counts()
```

```python
mapa_condicoes = {
    'Céu Claro': 'Sol',
    'céu claro': 'Sol',
    'Nublado': 'Sol',
    'nublado': 'Sol',
    'Chuva':'Chuva',
    'chuva':'Chuva',
    'Garoa/Chuvisco':'Chuva',
    'garoa/chuvisco':'Chuva',
    'Sol': 'Sol',
    'sol': 'Sol',
    'Nevoeiro/Neblina':'Neblina',
    'nevoeiro/neblina':'Neblina',
    'Vento': 'Sol',
    'vento': 'Sol',
    'Granizo':'Chuva',
    'granizo':'Chuva',
}
df['condicao_metereologica'] = df['condicao_metereologica'].astype(str).str.lower().replace(mapa_condicoes)
df = df[df['condicao_metereologica'].astype(str).str.lower() != 'nan']
df['condicao_metereologica'].value_counts()
```
