```python
import pandas as pd
from pycaret.classification import setup, compare_models, finalize_model, predict_model
```

```python
df = pd.read_csv('df_tratado.csv', sep=',')
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0', 'id', 'data_inversa', 'horario', 'tracado_via', 'latitude', 'longitude', 'regional', 'delegacia', 'uop', 'origem_arquivo', 'classificacao_acidente'])
```

```python
df.head()
```

```python
# Suponha que seu dataframe se chama df e já tem as colunas que listou
df = df.dropna()
categorical_cols = ['dia_semana', 'uf', 'br', 'municipio',
                    'causa_acidente', 'tipo_acidente',
                    'fase_dia', 'sentido_via', 'condicao_metereologica',
                    'tipo_pista', 'uso_solo']

numeric_cols = ['km', 'pessoas', 'feridos', 'veiculos']

clf1 = setup(data=df,
             target='grave',
             session_id=123,
             categorical_features=categorical_cols,
             numeric_features=numeric_cols,
             fix_imbalance=True,
             imputation_type='simple'
            )

best_model = compare_models()

final_model = finalize_model(best_model)

predictions = predict_model(final_model, data=df)

predictions.head()
```
