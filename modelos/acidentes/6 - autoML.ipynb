```python
import pandas as pd
from pycaret.classification import setup, compare_models, finalize_model, predict_model
```

```python
df = pd.read_csv('df_tratado.csv', sep=',')
df.head()
```

```python
df.columns
```

```python
# Suponha que seu dataframe se chama df e já tem as colunas que listou

categorical_cols = ['dia_semana', 'horario', 'uf', 'br', 'municipio',
                    'causa_acidente', 'tipo_acidente', 'classificacao_acidente',
                    'fase_dia', 'sentido_via', 'condicao_metereologica',
                    'tipo_pista', 'tracado_via', 'uso_solo', 'regional',
                    'delegacia', 'uop']

numeric_cols = ['km', 'pessoas', 'mortos', 'feridos_leves', 'feridos_graves',
                'ilesos', 'ignorados', 'feridos', 'veiculos', 'dia', 'mes', 'ano']

clf1 = setup(data=df,
             target='grave',
             session_id=123,
             categorical_features=categorical_cols,
             numeric_features=numeric_cols,
             fix_imbalance=True,
             imputation_type='simple'
            )

best_model = compare_models()

final_model = finalize_model(best_model)

predictions = predict_model(final_model, data=df)

print(predictions.head())
```
